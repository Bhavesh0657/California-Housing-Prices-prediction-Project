```python
# Q6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.
```

```python
import pandas as pd

class DatasetAnalyzer:
    def __init__(self, data):
        self.dataset = pd.DataFrame(data)

    def calculate_median(self, column_name):

        if column_name in self.dataset.columns:
            return self.dataset[column_name].median()
        else:
            return None

# Create an instance of DatasetAnalyzer
dataset_analyzer = DatasetAnalyzer(data)

# Calculate the median for total_bedroom
total_bedrooms = dataset_analyzer.calculate_median('total_bedrooms')
print(f"Median of total_bedroom: {total_bedrooms}")
```

```python
# Q10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

housing_data = pd.DataFrame(data)

# Define a function to categorize total bedrooms
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Create a new column 'total_bedroom_size' based on the categorization
housing_data['total_bedroom_size'] = housing_data['total_bedrooms'].apply(categorize_bedroom_size)

# Display the modified DataFrame
print(housing_data)
```

```python
Analysis:

# @ 'total_bedrooms' is a continuous variable as it can take on real values.
# -- after user define function (total_bedrooms is classifed or seperate in 3 parts accdoring to there quantity)
# @ 'total_bedroom_size' is now a categorical variable with three categories: 'small,' 'medium,' and 'large.'
```
